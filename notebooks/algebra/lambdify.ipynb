##### Lambdify
Sympy is great for symbolic computations (algebra, calculus), but it's slow for numerical computations.

`sympy.lambdify` converts a symbolic expression into a fast numerical function that we can use just like a regular python function, even with numpy

```python
import sympy as sp
from sympy.abc import x

expr   = x**2 + 3*x + 2
f_expr = sp.lambdify(x, expr)
f_expr(1e5)
```

If we want to pass our lambdify's expression into numpy, we need to do a little bit more work

```python
import numpy as np

expr   = sp.sin(x) + x**2
f_expr = sp.lambdify(x, expr, 'numpy')

x_vals = np.linspace(-5, 5, 100)
y_vals = f_expr(x_vals)

# now, we can also plot our function
import matplotlib.pyplot as plt
plt.plot(x_vals, y_vals)
```

We can also have multple variables:

```python
from sympy.abc import y

expr = x**2 + y**2 + x*y
f_expr = sp.lambdify((x,y), expr, 'numpy')
f_expr(5, 2)

# We can even pass our own dictionary or mix modules:

expr   = sp.sin(x) + sp.sin(y) + sp.sin(x)*sp.sin(y)
f_expr = sp.lambdify((x,y), expr, modules=['numpy', {'sin': np.sin}])

x_vals = np.linspace(0, 5, 100)
y_vals = f_expr(x_vals, x_vals*2)
plt.plot(x_vals, y_vals)
```
