```python
import matplotlib.pyplot as plt
import numpy as np
import scienceplots
plt.style.use(['science', 'notebook', 'no-latex', 'grid'])
```

#### 1. Line Plots
Simple plotting commands to look at data quickly

```python
x = np.linspace(0, 15, 30)
y = np.sin(x) + 0.25*np.random.randn(len(x))
```

```python
plt.figure(figsize=(10,3))
# Different styles are available x--, --, -, ^--, <--, >--, o, o--
plt.plot(x, y, 'o--', lw=1, color='purple', ms=5)
```

Axis Labels and Legends

```python
plt.figure(figsize=(10, 3))
plt.plot(x, y, 'o--', color='purple', lw=0.7, ms=6, label="capacitor voltage")
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend(loc='upper right', fontsize=10)
```

Multiple lines on the same plot

```python
x2 = np.linspace(0, 15, 100)
y2 = np.sin(x2)
```

```python
plt.figure(figsize=(10, 3))
plt.plot(x,   y, 'o--', color='purple', lw=0.7, ms=6, label="capacitor voltage")
plt.plot(x2, y2, '-', color='green', lw=0.7, ms=6, label="diode voltage")
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend(loc='upper right', fontsize=10)
```

$x$ and $y$ limits

```python
plt.figure(figsize=(10, 3))
plt.plot(x,   y, 'o--', color='purple', lw=0.7, ms=6, label="capacitor voltage")
plt.plot(x2, y2, '-', color='green', lw=0.7, ms=6, label="diode voltage")
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.ylim(top=2, bottom=-2)
plt.xlim((0, 15))
plt.legend(loc='upper right', fontsize=10, ncol=2)
```

#### 2. Histograms

```python
res = np.random.randn(1000)*0.2 + 0.4
```

Quick Histogram

```python
plt.figure(figsize=(6, 5))
plt.hist(res, bins=100, density=True)
```

Different style of bins (useful for plotting multiple histograms)

```python
res2 = np.random.randn(1000)*0.2 + 0.4

plt.figure(figsize=(5, 5))
plt.hist(res, bins=30, density=True, histtype='step')
plt.hist(res2, bins=30, density=True, histtype='step')
```

#### 3. A more Advanced API
For more complicated plots, such as:
- Multiple subplots on figure 1
- Conveniently adding text to plots
- Creating animations
We need to use a slightly more advanced API

##### Single Axes in one Figure

```python
fig, ax = plt.subplots(1, 1, figsize=(10,3))
ax.plot(x, y, 'o--')
ax.set_xlabel('asd')
ax.set_ylabel('asdasd')
fig.tight_layout()
```
