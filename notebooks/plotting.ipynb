#### 1. Line Plots
Simple plotting commands to look at data quickly

```python
x = np.linspace(0, 15, 30)
y = np.sin(x) + 0.25*np.random.randn(len(x))
```

```python
plt.figure(figsize=(10,3))
# Different styles are available x--, --, -, ^--, <--, >--, o, o--
plt.plot(x, y, 'o--', lw=1, color='purple', ms=5)
```

Axis Labels and Legends

```python
plt.figure(figsize=(10, 3))
plt.plot(x, y, 'o--', color='purple', lw=0.7, ms=6, label="capacitor voltage")
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend(loc='upper right', fontsize=10)
```

Multiple lines on the same plot

```python
x2 = np.linspace(0, 15, 100)
y2 = np.sin(x2)
```

```python
plt.figure(figsize=(10, 3))
plt.plot(x,   y, 'o--', color='purple', lw=0.7, ms=6, label="capacitor voltage")
plt.plot(x2, y2, '-', color='green', lw=0.7, ms=6, label="diode voltage")
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend(loc='upper right', fontsize=10)
```

$x$ and $y$ limits

```python
plt.figure(figsize=(10, 3))
plt.plot(x,   y, 'o--', color='purple', lw=0.7, ms=6, label="capacitor voltage")
plt.plot(x2, y2, '-', color='green', lw=0.7, ms=6, label="diode voltage")
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.ylim(top=2, bottom=-2)
plt.xlim((0, 15))
plt.legend(loc='upper right', fontsize=10, ncol=2)
```

##### More Complex Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
import scienceplots

plt.style.use(['science', 'grid', 'no-latex'])


x = np.linspace(0, 15, 30)
y = np.sin(x) + 0.25*np.random.randn(len(x))


# fig, ax = plt.subplots(1,1, figsize=(5,3), dpi=200)
fig, ax = plt.subplots(1,1, figsize=(10,3))
fig.set_facecolor('black')
ax.set_facecolor('black')
ax.grid(color='white', linewidth=0.5, alpha=0.6, zorder=0)
ax.plot(x, y, 'o--', color='orange', lw=1, ms=5)
ax.tick_params(axis='both', colors='white')
ax.set_xlabel("Time [s]", color='white')
ax.set_ylabel("Voltage [V]", color='white')
ax.set_ylim((-2, 2))
fig.tight_layout()

```

#### Gradient Line Plots

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

x = np.linspace(0, 10, 500)
y = np.sin(x)

# To make a gradient line plot, we want to break the line into small segements so we can color
# each one individually.
points = np.array([x, y]).T.reshape(-1,1,2)
segements = np.concatenate([points[:-1], points[1:]], axis=1)

colors = x[:-1]

lc = LineCollection(segements, cmap='rainbow', norm=plt.Normalize(x.min(), x.max()))
lc.set_array(colors)
lc.set_linewidth(2)

fig, ax = plt.subplots()
ax.add_collection(lc)
ax.set_xlim((x.min(), x.max()))
ax.set_ylim((y.min(), y.max()))

```

#### Making Animations
`matplotlib.animination.FuncAnimation(fig, func, frames, init_func, fargs, interval)`

1. Relevant Inputs:
- `fig`: The figure we want to animate
- `func`: A function that updates our data (func(frame, *fargs))
- `frames`: Total number of frames
- `init func`: Initial plot
- `fargs`: additional arguments
- `interval`: Delay between each frame in [ms]

2. Optional Args
- `blit`(bool): Optimizing drawing

3. Methods
- `pause()`
- `resume()`
- `save()`

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np

t = np.linspace(0, 10, 100)
y = np.sin(t)

fig, axis = plt.subplots(1,1)
axis.set_xlim((np.min(t), np.max(t)))
axis.set_ylim((np.min(y), np.max(y)))

animated_plot, = axis.plot([],[])

def update(frame):
    animated_plot.set_data(t[: frame], y[: frame])
    return animated_plot

animation = FuncAnimation(fig=fig, func=update, frames=len(t), interval=25)

```
